### Tensor datatypes

**Note:** Tensor datatypes is one of the 3 big errors you'll run into with PyTorch & deep learning:

1. Tensors not right `datatype`
2. Tensors not right `shape`
3. Tensors not on the right `device`

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                              dtype=None, # defines tensors data type
                              device=None, # tensors on which device runnin on
                              requires_grad=False) #  whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
# Float 32 tensor
float_16_tensor = torch.tensor([3.0, 6.0, 9.0],
                              dtype=torch.float16, # defines tensors data type
                              device=None, # tensors on which device runnin on
                              requires_grad=False) #  whether or not to track gradients with this tensors operations
float_16_tensor
```

```python
# convert 32 bytes into 16 bytes
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
# operation with 16 bytes and 32 bytes of data type
float_16_tensor * float_32_tensor
```

```python
# it woks with these bytes
float_32_tensor * float_16_tensor
```

```python
float_32_tensor * float_32_tensor
```

```python
float_16_tensor * float_16_tensor
```
