### Tensor datatypes

**Note:** Tensor datatypes is one of the 3 big errors you'll run into with PyTorch & deep learning:

1. Tensors not right `datatype`
2. Tensors not right `shape`
3. Tensors not on the right `device`

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                              dtype=None, # defines tensors data type
                              device=None, # tensors on which device runnin on
                              requires_grad=False) #  whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
# Float 32 tensor
float_16_tensor = torch.tensor([3.0, 6.0, 9.0],
                              dtype=torch.float16, # defines tensors data type
                              device=None, # tensors on which device runnin on
                              requires_grad=False) #  whether or not to track gradients with this tensors operations
float_16_tensor
```

```python
# convert 32 bytes into 16 bytes
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
# operation with 16 bytes and 32 bytes of data type
float_16_tensor * float_32_tensor
```

```python
# it woks with these bytes
float_32_tensor * float_16_tensor
```

```python
float_32_tensor * float_32_tensor
```

```python
float_16_tensor * float_16_tensor
```

### Reshaping, stacking, squeezing and unsqueezing tensors

```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
x.shape
```

```python
x.reshape(3, 3)
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
x.view(3, 3)
```

```python
z = x.view(3, 3)
z, z.shape
```

```python
z[:, 0]
```

```python
z = z.view(1, 9)
z
```

```python
# update the value
z[:, 0] = 5
```

```python
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked, x_stacked.shape
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked, x_stacked.shape
```

```python
x_stacked.shape
```

```python
x_reshaped
```

```python
# remove one dimention from the tensor
x_squeezed = x_reshaped.squeeze()
x_squeezed, x_squeezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed, x_unsqueezed.shape
```
